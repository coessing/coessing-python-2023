## Exercice 1

Écrivez une fonction qui, en fonction d'un chemin de fichier qui contient un poème, numérotera les lignes du poèmes suivi d'une tabulation (marquée par "\t") et retournera cette nouvelle version du texte.

Les lignes vides ne seront pas numérotées.

**Aide** : 
- `.split()` peut prendre un argument correspondant à la valeur sur laquelle découper un texte. `.split('\n')` découpera un texte sur les nouvelles lignes

````python
def numeroter_poeme(chemin_de_fichier):
    """ Numérote un poème
    
    :param chemin_de_fichier: Chemin du fichier texte contenant le poème
    :type chemin_de_fichier: str
    :returns: Liste de vers où le numéro de ligne est indiqué 
              à gauche avant une tabulation 
    :rtype: list
    """
    # On ouvre le fichier et on stocke son "lecteur" dans fichier_ouvert
    with open(chemin_de_fichier) as fichier_ouvert:
        # On lit le fichier et on stocke sa valeur dans texte
        texte = fichier_ouvert.read()
    # On crée une variable d'index qui nous servira à compter les lignes
    numero_vers = 1
    # On crée une variable qui nous permet de stocker les vers dans une liste
    poeme = []
    # on boucle sur les lignes découpées d'après .split("\n")
    for line in texte.split("\n"):
        # Line est désormais équivalent à une des lignes du fichier original
        # pour vérifier, vous pouvez décommentez la ligne suivante
        # print(line) 
        
        # Il est possible qu'une ligne soit vide.
        # Une ligne vide est ne vaut pas True (boolean("") = False )
        # On peut donc faire un if 
        if line:
            # Pour concaténer un nombre à une chaîne, on doit le typer/caser.
            # On utiliser donc str()
            poeme.append(str(numero_vers)+"\t"+line)

            # On incrémente (augmente de 1) notre numéro de vers pour qu'au prochain vers
            # le numéro corresponde à la bonne ligne
            numero_vers += 1
    
    # Une fois la boucle finie (et donc hors de l'indentation de la boucle)
    # on s'assure de retourner la valeur
    return poeme

# Code pour tester le résultat
poeme = numeroter_poeme("data/lanuitdelamort.michel.txt")

# Pour débugguer
print(poeme)
# Pour tester
assert "5\tJamais, jamais le dernier jour" in poeme
assert "13\tElle monte et les couvrira." in poeme
````

**Aide** 
- une ligne vide est équivalent à `False` 

````python
# Cet exercice est une variante un peu plus complexe des exercices précédents
# Il s'agira de faire des sous listes contenant des strophes.
# Les strophes sont indiquées par une/des ligne(s) vide(s) séparant des lignes pleines
# Il faudra donc, lors d'un changement de ligne, changer potentiellement de sous-liste
def numeroter_poeme_avec_strophe(chemin):
    """ Numérote un poème avec des sous-listes représentant les strophes
    
    :param chemin: Chemin du fichier texte contenant le poème
    :type chemin: str
    :returns: Liste de vers où le numéro de ligne est indiqué 
              à gauche avant une tabulation 
    :rtype: list
    """
    # On ouvre le fichier et on stocke son "lecteur" dans fichier_ouvert
    with open(chemin) as fichier_ouvert:
        # On lit le fichier et on stocke sa valeur dans texte
        texte = fichier_ouvert.read()
    # On crée une variable d'index qui nous servira à compter les lignes
    numero_vers = 1
    # On crée une variable qui nous permet de stocker les vers dans une liste
    # Puisque nous commencerons par une ligne pleine
    # Nous créons directement la strophe 
    # Les premiers crochets (poeme = [  ]) représentent le poème
    # les deuxièmes ( poeme =  [] ) représentent la strophe
    poeme = [[]]
    
    # on boucle sur les lignes découpées d'après .split("\n")
    for line in texte.split("\n"):
        # Line est désormais équivalent à une des lignes du fichier original
        # pour vérifier, vous pouvez décommentez la ligne suivante
        # print(line) 
        
        # Il est possible qu'une ligne soit vide.
        # Une ligne vide est ne vaut pas True (boolean("") = False )
        # On peut donc faire un if 
        if line:
            # Pour écrire un tuple, on met entre parenthèse et on joint par des virgules
            # On ajoute ce poème à la dernière strophe ouverte
            # Notre poème étant une liste faite de sous-listes représentant les strophes,
            # on accède à la dernière strophe/liste via poeme[-1]
            poeme[-1].append((numero_vers, line))

            # On incrémente (augmente de 1) notre numéro de vers pour qu'au prochain vers
            # le numéro corresponde à la bonne ligne
            numero_vers += 1
            
        # Si on a pas de ligne pleine, on est entre deux strophes
        # C'est à ce moment là qu'il va falloir créer une nouvelle strophe en prévision
        # de la prochaine ligne pleine.
        else:
            # Il est imaginable que pour la clarté du texte, on ait deux lignes vides
            # Dans ce cas, on va vérifier que notre dernière strophe a bien été remplie
            # Avant de créer une autre ligne vide
            if len(poeme[-1]):
                # Pour créer une nouvelle strophe,
                # On ajoute une liste au poeme
                poeme.append([])
    
    # Une fois la boucle finie (et donc hors de l'indentation de la boucle)
    # on s'assure de retourner la valeur
    return poeme
        
poeme = numeroter_poeme_avec_strophe("data/Ballade.XXVIII.dePisan.txt")

# On débuggue
print(poeme)

# On vérifie
assert poeme[0] == [(1, 'Mon doulz ami, vueilliez moy pardonner,'),
  (2, 'Se je ne puis, si tost com je vouldroye,'),
  (3, 'Parler a vous, car ainçois ordener'),
  (4, 'Me fault comment sera, ne par quel voye.'),
  (5, 'Car mesdisans me vont gaitant'),
  (6, 'Qui du meschief et du mal me font tant,'),
  (7, 'Que je ne puis joye ne bien avoir,'),
  (8, 'Pour le desir que j’ay de vous veoir.')]
assert poeme[-1][0] == (17, 'Mais ne sçaront ja eulx si fort pener,')
````

Le fichier data/debat.folieamour.labbe.txt contient un dialogue entre Folie et Amour écrit par Louise Labbé. Vous produirez à partir de ce fichier une liste de réplique et de didascalie de telle sorte que :

- les répliques de personnages soient des `tuple` prenant la forme `(index_de_ligne, "réplique", nom_de_personnage, texte de la replique)`
- les didascalies soient des tuples prenant la forme `(index_de_ligne, "didascalie", False, texte de la didascalie)`
- les lignes vides soient ignorées

````python
def debat(chemin):
    """ Numérote un débat en stockant les informations nécessaires
    
    Les fichiers respectent la forme suivante :
    
    - Une réplique == Une ligne
    - Il peut y avoir des lignes vides entre les répliques
    - Une réplique commence par un nom de personnage en Majuscule
    - Si il est en minuscule, il s'agit d'une didascalie
    
    :param chemin: Chemin du fichier texte contenant le débat
    :type chemin: str
    :returns: Une liste de tuple permettant d'identifier un numéro de ligne, le type de ligne, 
    le nom de l'intervenant le cas échéant et enfin le texte de la ligne défait de ses métadonnées.
    :rtype: list
    """
    # On ouvre le fichier et on stocke son "lecteur" dans fichier_ouvert
    with open(chemin) as fichier_ouvert:
        # On lit le fichier et on stocke sa valeur dans texte
        texte = fichier_ouvert.read()
        
    # On crée une variable d'index qui nous servira à compter les lignes
    num_ligne = 1
    # On crée une variable qui nous permet de stocker les lignes dans une liste
    texte_parse = []
    
    # on boucle sur les lignes découpées d'après .split("\n")
    for line in texte.split("\n"):
        # Line est désormais équivalent à une des lignes du fichier original
        # pour vérifier, vous pouvez décommentez la ligne suivante
        # print(line) 
        
        # Il est possible qu'une ligne soit vide.
        # Une ligne vide est ne vaut pas True (boolean("") = False )
        # On peut donc faire un if 
        if line:
            # Le premier mot d'une ligne nous informe du type de ligne
            # Si il commence par FOLIE. ou AMOUR., la ligne est un dialogue
            # Sinon, c'est une didascale
            # Note : il y a des façons plus "intelligentes" de faire cette partie,
            #        notamment à partir d'expression régulière. Il faudra attendre 
            #        le dit chapitre pour les voir
            if line.startswith("FOLIE."):
                # On va chercher à enlever l'information de personnage de la ligne
                # Pour faire cela, on va remplacer l'information
                propre = line.replace("FOLIE.", "")
                # str.strip() permet d'enlever les espaces en début et fin de ligne
                # afin de normaliser celle-ci
                propre = propre.strip()
                texte_parse.append(
                    (num_ligne, "réplique", "FOLIE", line.replace("FOLIE.", "").strip())
                )
            elif line.startswith("AMOUR."):
                # On applique la même logique qur sur le if précédent
                # en évitant de créer de nouvelles variables
                texte_parse.append(
                    (num_ligne, "réplique", "AMOUR", line.replace("AMOUR.", "").strip())
                )
            else:
                # On a une didascalie 
                # On ajoute simplement la ligne au texte
                texte_parse.append(
                    (num_ligne, "didascalie", False, line)
                )
                

            # On incrémente (augmente de 1) notre numéro de vers pour qu'au prochain vers
            # le numéro corresponde à la bonne ligne
            num_ligne += 1
    
    # Une fois la boucle finie (et donc hors de l'indentation de la boucle)
    # on s'assure de retourner la valeur
    return texte_parse
    
# Tests
resultats = debat("data/debat.folieamour.labbe.txt")
assert isinstance(resultats, list) == True, "Le resultat doit être une liste"
assert resultats[3] == (
    4,
    "réplique",
    "AMOUR",
    "II n'en ira pas ainsi : car avant que tu m'eschapes, je te donneray à connoitre que tu ne te dois atacher à moy."
)
assert resultats[16] == (
    17,
    "didascalie",
    False,
    "Folie se fait invisible, tellement, qu'Amour ne la peut assener."
)
assert resultats[8] == (
    9,
    "réplique",
    "FOLIE",
    "Tu es Amour, fils de Venus."
)
````

# Bonus : d'une fonction qui "parse" un texte à une fonction qui écrit du xml

La fonction `numeroter_poeme_avec_strophe` permet de faire d'un fichier texte un ensemble de fichiers python. Il est encore plus intéressant de voir ce que l'on peut faire avec cela :

````python
# Pensez à exécuter la cellule qui contient la définition de la fonction numeroter_poeme_avec_strophe
poeme = numeroter_poeme_avec_strophe("data/Ballade.XXVIII.dePisan.txt")

def poeme_strophe_to_tei(poeme, titre, auteur, chemin_template_tei="data/xml/template_tei_exercice.xml"):
    """ Transforme un poeme sous la forme d'une liste de listes de vers en un texte de TEI
    
    Le template est un fichier XML où je dois pouvoir trouver les chaînes:
    
    - $titre$ que je remplacerai par le titre fourni
    - $auteur$ que je remplacerai par l'auteur-e fourni-e
    - $poeme$ que je remplacerai par le XML de mon poeme
    
    :param poeme: Contenu du poeme
    :type poeme: list
    :param titre: Titre du poeme
    :type titre: str
    :param auteur: Nom de l'auteur
    :type auteur: str
    :param chemin_template_tei: (Optionnel) Chemin du template
    :type chemin_template_tei: str
    :returns: Poeme en tei sous la forme d'une chaine de caractères
    :rtype: str
    """
    # Je vais utiliser les guillemets simples afin de ne pas avoir à échapper constamment
    # les guillemets doubles
    # J'ouvre une div poeme (je finis ma ligne par un saut de ligne)
    tei_poeme = '<div type="textpart" subtype="poeme">\n'
    
    # Je vais boucler sur mes strophes
    for strophe in poeme:
        # A ma première itération, strophe est égal à poeme[0]
        # A ma deuxième, strophe est égal à poeme[1]
        # etc. jusqu'à ce qu'il n'y ait plus d'élément dans poeme
        
        # Une strophe est marquée par un <lg> en TEI
        # Je concatène ma chaine représentant du XML à ma chaine précédente
        tei_poeme += "<lg>\n"
        
        # Mes vers font partie de ma strophe sous la forme d'une liste
        # Ces vers sont des tuples : je peux les dézipper/unziper/unpacker
        # Pour ce faire, je sépare directement ma variable-tuple en deux variable
        for numero_vers, vers in strophe:
            # C'est équivalent à :
            # for tuple_de_vers in strophe:
            #     numero_vers = tuple_de_vers[0]
            #     vers = tuple_de_vers[1]
            
            # Ici j'écris mon vers dans la tei de mon poème
            # Mon vers est constitué de <l> avec un attribut n et du texte
            # Le numero de vers étant un nombre, je dois le caster
            tei_poeme += '<l n="'+str(numero_vers)+'">'+vers+'</l>\n'
        
        # Je ferme ma strophe et sautre une ligne
        tei_poeme += "</lg>\n"
        
    # Je ferme enfin mon poeme
    tei_poeme += "</div>\n"
    
    # Je lis mon template et je stock son lecteur dans lecteur_template
    with open(chemin_template_tei) as lecteur_template:
        # Je lis et je stocke le contenu du template dans la variable xml_final
        xml_final = lecteur_template.read()
        
    # Je remplace le titre
    xml_final = xml_final.replace("$titre$", titre)
    
    # Je remplace l'auteur
    xml_final = xml_final.replace("$auteur$", auteur)
    
    # Je remplace le poeme
    xml_final = xml_final.replace("$poeme$", tei_poeme)
    
    # Je retourne le xml_final
    return xml_final

tei = poeme_strophe_to_tei(poeme, "Ballade XXVIII", "Christine de Pisan")

# Et on affiche le résultat
print(tei)
# (On aurait pu aussi l'enregistrer après l'avoir afficher...)
````
