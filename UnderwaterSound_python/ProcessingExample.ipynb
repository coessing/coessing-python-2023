Acoustic Data Processing

We first read in several libraries or packages. The first thing we read in is file DataProcessingLib.py, which is in this folder and contains several acoustic processing functions. Another library we will be using is called pydub, it allows us to work with .wav files.  This is a common format for acoustic data files.

```python
import DataProcessingLib as lib
from scipy.io import wavfile
import scipy.io
import matplotlib.pyplot as plt
import numpy as np
from pydub import AudioSegment
```

We'll start by exploring acoustic data from the DOSITS website.  The file 'LargeCommercialShip1.wav' is a short recording of a large ship driving by.  

The line below reads the timeseries into 'data' and reads in the sampling frequency (fs).
Let's also print out the sampling frequency of this file. This is given in units of samples per second or Hertz (Hz).

```python
fs,data=wavfile.read('WaveBeach.wav')  # scipy.io.wavfile.read: read .wav file into sampling frequency and data points
fs
```

We can plot the timeseries which shows the amplitude of the acousitc signal as a function of time.  We need to include the sampling frequency to get the correct time axis.

```python
lib.waveform(data, fs, title = 'Acoustic Data Time Series')  # plot waveform 
```

The timeseries shows how the amplitude changes with time, but does not tell us much about the frequency content of the signal.  A spectrogram shows how the frequency content of a signal changes over the duration of the signal. 

```python
lib.spectrogram(data, fs, title='Spectrogram')  # plot spectrogram
```

Let's try this with a marine mammal vocalization.  

```python
fs,data=wavfile.read('PilotWhale.wav')  
lib.waveform(data, fs, title = 'Acoustic Data Time Series - Pilot Whale') 
```

```python
lib.spectrogram(data, fs, title='Spectrogram - Pilot Whale')  
```

Next we will read in acoustic data recorded data using a package from Cetacean Research Technology.  This package includes a S26-H1 Hydrophone (a hydrophone is an underwater microphone) and a H1n acoustic data recorder.  The data recorder records acoustic signals in stereo, meaning there are 2 channels.  We read in the wav file below and then reduce it to a single channel ('mono') to work with the data.  

Put your wav file in your data processing directory and enter the filename below. 

```python
filename = ('230807-125342.wav')      #Enter wav file to process here
sound = AudioSegment.from_wav(filename)
sound = sound.set_channels(1)
sound.export('mono.wav', format = 'wav')

fs, data = wavfile.read('mono.wav')  
lib.waveform(data, fs, title = 'Acoustic Data Time Series')  # plot waveform 16-bit integer PCM
```

```python
lib.spectrogram(data, fs, title='Recorded Spectrogram')  # plot spectrogram of unfiltered data
```
